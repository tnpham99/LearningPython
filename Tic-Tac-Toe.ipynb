```python
board = [['-','-','-'],
         ['-','-','-'],
         ['-','-','-']]

board_avail = ['00','01','02','10','11','12','20','21','22']

def main():
    print('Welcome to the traditional Tic Tac Toe game. Please choose from the menu:'+'\n'+
         '1. Start game'+'\n'+
         '2. Game introduction and instruction')
    user_menu = int(input())
    if user_menu == 1:
        game()
    elif user_menu == 2:
        intro()
    else:
        print('Invalid choice.')
        main()
    
def intro():
    print('Tic-tac-toe, noughts and crosses, or Xs and Os is a game for two players (player and CPU in this case) who take turns marking the spaces in a three-by-three grid with X or O (below).'+'\n'+
          '   0 | 1 | 2'+'\n'+
          '0 ___|___|___'+'\n'+
          '1 ___|___|___'+'\n'+
          '2    |   |   '+'\n'+
          '\n'+
          'The player who succeeds in placing three of their marks in a horizontal, vertical, or diagonal row is the winner.'+'\n'+
          '\n'+
          'In this version of Tic Tac Toe, you will make your moves by typing the location of the square you want to mark in the order of row to column.'+'\n'+
          'For example: Bottom left square is located at row 2 and column 0, you will then type in when asked row: 2 and column: 0.'+'\n'+
          '\n'+
          'Are you ready?')
    user_intro = input('YES or NO: ')
    if user_intro == 'YES':
        game()
    elif user_intro == 'NO':
        print('Come back and play when you are ready!')
    else:
        print('invalid answer.')
        intro()

#The main part
def game():
    import random
    
    count = 0
    while count < 9:
        if board[0][0] == board[0][1] == board[0][2] == 'X' or board[1][0] == board[1][1] == board[1][2] == 'X' or board[2][0] == board[2][1] == board[2][2] == 'X' or board[0][0] == board[1][0] == board[2][0] == 'X' or board[0][1] == board[1][1] == board[2][1] == 'X' or board[0][2] == board[1][2] == board[2][2] == 'X' or board[0][0] == board[1][1] == board[2][2] == 'X' or board[0][2] == board[1][1] == board[2][0] == 'X':
            print(*board, sep='\n')
            print('You win! :)')
            break
        elif board[0][0] == board[0][1] == board[0][2] == 'O' or board[1][0] == board[1][1] == board[1][2] == 'O' or board[2][0] == board[2][1] == board[2][2] == 'O' or board[0][0] == board[1][0] == board[2][0] == 'O' or board[0][1] == board[1][1] == board[2][1] == 'O' or board[0][2] == board[1][2] == board[2][2] == 'O' or board[0][0] == board[1][1] == board[2][2] == 'O' or board[0][2] == board[1][1] == board[2][0] == 'O':
            print(*board, sep='\n')
            print('Oh no, CPU wins :(')
            break
        else:
            if count%2 == 0:
                print('Choose a square you want to mark: ')
                row = int(input('row: '))
                column = int(input('column: '))
                if board[row][column] == '-':
                    board[row][column] = 'X'
                    count += 1
                    board_avail.remove(str(row)+str(column))
                else:
                    print('Square is not available. Choose a different square.')
            else:
                square = random.choice(board_avail)
                board[int(square[0])][int(square[1])] = 'O'
                count += 1
                print(*board, sep='\n')
            print(board_avail)
    else:
        print(*board, sep='\n')
        print("It's a draw!")
main()
```
